```python
#Openning the csv and taking a look at what are we working with.
df = pd.read_csv("../attacks.csv")
df.head()
```

```python
#getting rid of NaN rows and the empty columns

df_clean = df.dropna(thresh= 3).drop(['Unnamed: 22','Unnamed: 23'], axis=1)
df_clean
```

### AGE

```python
df_clean['Age'].value_counts(dropna = False)
```

```python
df_clean['Age'].unique()
```

```python
df_clean['Age'] = df_clean['Age'].astype(str)
df_clean['Age'] = df_clean['Age'].str.replace(r'(^.*een.*$)', '16')
df_clean['Age'] = df_clean['Age'].str.replace(r'(^.*oun.*$)', '25')
df_clean['Age'] = df_clean['Age'].str.replace(r'(^.*dul.*$)', '45')
df_clean['Age'] = df_clean['Age'].str.replace(r'(^.*lderl.*$)', '70')
df_clean['Age'] = df_clean['Age'].str.replace(r'(^.*middle-a.*$)', '35')
```

```python
df_clean['Age'].unique()
```

```python
#Statistics
df_clean['Age'] = pd.to_numeric(df_clean['Age'])

print('Mean: ', df_clean['Age'].mean())
print('Median: ', df_clean['Age'].median())
print('Mode: ', df_clean['Age'].mode())
```
